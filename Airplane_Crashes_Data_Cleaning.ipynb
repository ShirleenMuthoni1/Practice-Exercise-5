```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('Airplane_Crashes_and_Fatalities_Since_1908.csv')
```

```python
# Question 5: Create a new dataframe from the raw/uncleaned data called ‘fatality_locations’ with the following columns: ‘Date’, ‘Location’, ‘Aboard’, ‘Fatalities’
fatality_locations = df[['Date', 'Location', 'Aboard', 'Fatalities']].copy()
print("New dataframe created with columns: Date, Location, Aboard, Fatalities")
print(f"Shape: {fatality_locations.shape}")
display(fatality_locations.head())
```

```python
# Question 8: Split Location column into Region and State/Country
print("Question 8: Splitting Location Column")
# Create a copy to avoid modifying original
fatality_locations_split = fatality_locations.copy()

# Split the Location column
split_locations = fatality_locations_split['Location'].str.split(', ', n=1, expand=True)

# Assign the split parts to new columns
fatality_locations_split['Region'] = split_locations[0]
fatality_locations_split['State_Country'] = split_locations[1]
# Display the result
print("Location column split into 'Region' and 'State_Country'")
display(fatality_locations_split[['Location', 'Region', 'State_Country']].head(10))
print("\n" + "="*50 + "\n")
```

```python
# Question 9: Order by fatalities (highest to lowest) and select first 100
print("Question 9: Top 100 Crashes by Fatalities")
top_100_fatalities = fatality_locations_split.sort_values('Fatalities', ascending=False).head(100)
print("Dataframe ordered by fatalities (highest to lowest) - Top 100:")
display(top_100_fatalities.head(10))  # Show first 10 of the top 100
print(f"\nTotal rows in top 100 dataframe: {len(top_100_fatalities)}")
print("\n" + "="*50 + "\n")
```

```python
# Question 10:  Generate a pie chart that shows the distribution of the top 25 number of fatalities per country/U.S. State. [Hint: charts can be generated from the seaborn library]
print("Question 10: Generating Pie Chart using Seaborn library - Top 25 Fatalities by Country/State")

# Group by State_Country and sum fatalities, then take top 25
country_fatalities = top_100_fatalities.groupby('State_Country')['Fatalities'].sum().sort_values(ascending=False)
top_25_countries = country_fatalities.head(25)

print(f"Top 25 countries/states by fatalities:")
for i, (country, fatalities) in enumerate(top_25_countries.items(), 1):
    print(f"{i:2d}. {country}: {fatalities} fatalities")

# Create pie chart using seaborn library
plt.figure(figsize=(14, 12))
colors = sns.color_palette('husl', len(top_25_countries))

plt.pie(top_25_countries.values,
        labels=top_25_countries.index,
        colors=colors,
        autopct='%1.1f%%',
        startangle=90,
        textprops={'fontsize': 9})

plt.title('Distribution of Top 25 Number of Fatalities by Country/U.S. State',
          fontsize=16, fontweight='bold', pad=20)
plt.axis('equal')

plt.tight_layout()
plt.show()

print("Pie chart created successfully!")
```
